```python
#Relabel Gender
#0 - Male
#1 - Female
#-1 - Unknown

#Relatable Misdimeanor/Felony
#0 - Misdimeanor
#1 - Felony
#-1 - Neither

s,f = np.shape(df)
for i in range(s):
    if df.loc[i, 'Gender'] == 'M':
        df.loc[i, 'Gender'] = 0
    elif df.loc[i, 'Gender'] == 'F':
        df.loc[i, 'Gender'] = 1
    else:
        df.loc[i, 'Gender'] = -1

for i in range(s):
    if df.loc[i, 'M/F'] == 'M':
        df.loc[i, 'M/F'] = 0
    elif df.loc[i, 'M/F'] == 'F':
        df.loc[i, 'M/F'] = 1
    else:
        df.loc[i, 'M/F'] = -1        
df.head()

```

```python
dfNew = df[df.Gender != -1]
dfNew.head()
```

```python
Models = dfNew[['Age','Gender','M/F']]
Models.head()
Models = Models.dropna(axis=0,how='any')
X = Models.drop(['M/F'],axis=1)
y = Models.drop(['Age','Gender'],axis=1)

Models
```

```python
import numpy as np
import sklearn.model_selection as ms

#Returns scores from KFold cross validation of n folds
#X_train - training data frame
#y_train - traininng labels data frame
#clf - Classifier
#n - Number of folds

def cross_val_scores(X_train, y_train, clf=None, n=5):
    scores = []
    kf = ms.KFold(n, shuffle=True)
    for train, validate in kf.split(X_train):
        clf.fit(X_train[train], y_train[train])
        scores.append(clf.score(X_train[validate], y_train[validate]))
    return scores
```

```python
from sklearn.linear_model import LogisticRegression
X_train, x_test, y_train, y_test = ms.train_test_split(X,y,test_size=0.25)

print(X_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
logreg.score(X_train,y_train)
# print(logreg_scores)
```
